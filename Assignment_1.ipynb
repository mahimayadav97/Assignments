```python
### Assignment 1:

x = [1,2,3,4,5]
y = [0,2,4,6,8]
import matplotlib.pyplot as plt
## Resize the graph
plt.figure(figsize=(6,5), dpi=100)
## Plot a line graph
plt.plot(x,y,label='random', color='blue', linewidth=2)
## Add a title, labels, legend to your graph
plt.title('Assignment 1 : Line chart')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.show()
## Save the graph
plt.savefig('mygraph.png', dpi=300)
```
